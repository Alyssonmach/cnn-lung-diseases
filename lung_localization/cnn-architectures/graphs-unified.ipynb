```python
densenet_iou = np.loadtxt('/content/gdrive/MyDrive/lung-localization/densenet/iou_mean_densenet.txt')
xception_iou = np.loadtxt('/content/gdrive/MyDrive/lung-localization/xception/iou_mean_xception.txt')
resnet_iou = np.loadtxt('/content/gdrive/MyDrive/lung-localization/resnet/iou_mean_resnet.txt')
mobilenet_iou = np.loadtxt('/content/gdrive/MyDrive/lung-localization/mobilenet/iou_mean_mobilenet.txt')
vgg_iou = np.loadtxt('/content/gdrive/MyDrive/lung-localization/vgg/iou_mean_vgg.txt')
inceptionresnet_iou = np.loadtxt('/content/gdrive/MyDrive/lung-localization/inceptionresnet/iou_mean_inceptionresnet.txt')
inception_iou = np.loadtxt('/content/gdrive/MyDrive/lung-localization/inception/iou_mean_inception.txt')
```

```python
plt.figure(figsize = (15, 8))
x_labels = ['Quinto Bloco', 'Quarto Bloco', 'Terceiro Bloco', 'Segundo Bloco', 'Primeiro Bloco']
plt.plot(x_labels, np.array(vgg_iou), label = 'VGG', linewidth = 2,  marker = '<', linestyle="--")
plt.plot(x_labels, np.array(resnet_iou), label = 'ResNet', linewidth = 3, marker = 's', linestyle=":")
plt.plot(x_labels, np.array(densenet_iou), label = 'DenseNet', linewidth = 3, marker="o", linestyle="--")
plt.plot(x_labels, np.array(mobilenet_iou), label = 'MobileNet', linewidth = 3, marker = '^', linestyle="-.")
plt.plot(x_labels, np.array(inception_iou), label = 'Inception', linewidth = 2, marker = '.', linestyle=":")
plt.plot(x_labels, np.array(inceptionresnet_iou), label = 'InceptionResNet', linewidth = 2, linestyle="-.")
plt.plot(x_labels, np.array(xception_iou), label = 'Xception', linewidth = 3, marker = '*',  linestyle="-.")
plt.xticks(size = 16, color = 'black')
plt.yticks(color = 'black')
plt.title('Congelamento de Blocos das Arquiteturas', size = 16, color = 'blue')
plt.xlabel('Blocos da Arquiteturas', size = 14, color = 'blue')
plt.ylabel('Inteserction over Union', size = 14, color = 'blue')
plt.grid(True)
plt.legend(prop={'size': 15})
plt.ylim([0.82,0.92])
```
